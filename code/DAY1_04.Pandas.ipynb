## 1. 데이터 타입
### (1) 데이터프레임
- R의 데이터프레임 타입을 모방
- 내부적으로는 numpy array를 참조하므로 배열연산이 가능하다.
- 2차원 행렬
- index, column

### (2) 시리즈
- 시리즈 클래스는 NumPy에서 제공하는 1차원 배열과 비슷하지만 각 데이터의 의미를 표시하는 인덱스(index)를 붙일 수 있다. 데이터 자체는 값(value)라고 한다.

## 2. 데이터 프레임 생성

```python
## 딕셔너리로 데이터프레임을 만드는 방식
## 딕셔너리의 각 키 값은 column 으로 사용된다.

data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}
columns = ["지역", "2015", "2010", "2005", "2000", "2010-2015 증가율"]
index = ["서울", "부산", "인천", "대구"]
df = pd.DataFrame(data, index=index, columns=columns)
# df = pd.DataFrame(data, index=index)
df
```

```python
## 리스트로 데이터프레임을 만드는 방식

data = [
    ['수도권', 9904312, 9631482, 9762546, 9853972, 0.0283],
    ['경상권', 3448737, 3393191, 3512547, 3655437, 0.0163],
    ['수도권', 2890451, 2632035, 2517680, 2466338, 0.0982],
    ['경상권', 2466052, 2431774, 2456016, 2473990, 0.0141]
       ]

columns = ["지역", "2015", "2010", "2005", "2000", "2010-2015 증가율"]
index = ["서울", "부산", "인천", "대구"]
df = pd.DataFrame(data, 
                  index=index, 
                  columns=columns)
df
```

```python
## 데이터프레임의 데이터를 array 형태로 볼 수 있다.

df.values
```

## 6. 그룹 분석
- groupby
- pivot

### (1) groupby


```python
## 특정 컬럼의 값에 그룹 연산을 한다.

grouped = titanic.groupby(by = 'pclass')
```

```python
## 객실 클래스별 탑승객 나이 평균

grouped.age.mean()
```

```python
## 객실 클래스별, 성별 나이 평균

titanic.groupby(by = ['pclass', 'sex']).age.mean()
```

```python
## 객실 클래스별, 성별 나이 평균과 표준편차, 최고령자

titanic.groupby(['pclass', 'sex']).agg({'age'  : ['mean', 'std', 'max']})
```

```python
## 객실 클래스별, 성별 기술통계

titanic.groupby(['pclass', 'sex']).describe().T
```

### (2) pivot
- 엑셀의 피봇과 동일하다.
- 여러 개의 key값에 대한 그룹화 연산을 시행한다.
- pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')
    - data: 분석할 데이터프레임 (메서드일 때는 필요하지 않음)
    - values: 분석할 데이터프레임에서 분석할 열
    - index: 행 인덱스로 들어갈 키 열 또는 키 열의 리스트
    - columns: 열 인덱스로 들어갈 키 열 또는 키 열의 리스트
    - aggfunc: 분석 메서드
    - fill_value: NaN 대체 값
    - margins: 오른쪽과 아래에 합계를 붙일지 여부
    - margins_name: 합계 열(행)의 이름

```python
titanic.head()
```

```python
## 승선지와 객실 클래스별 평균 요금
## 두개의 key값으로 그룹화 연산을 한 것과 동일하다.

titanic.pivot_table(index = ['embark_town', 'sex'], columns = 'pclass', values = 'fare', aggfunc='mean')
```

```python
## 승선지와 객실클래스별, 성별 생존율을 구해보세요.
## key값이 3개임.

titanic.pivot_table(index = ['embark_town', 'sex'], columns = 'pclass', values = 'survived', aggfunc='mean')
```

```python
## 승선지와 객실클래스별, 성별 생존율을 구해보세요.
## key값이 3개임.

titanic.pivot_table(index = ['embark_town', 'sex'], columns = 'pclass', values = 'survived', aggfunc='mean')
```
